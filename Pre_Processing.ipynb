```python
import re
```

```python
import pandas as pd
```

```python
f=open('WhatsApp Chat with BS (DS) .txt','r',encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s[APMapm]{2}\s-\s'
```

```python
messages=re.split(pattern,data)[1:]
len(messages)
```

```python
dates=re.findall(pattern,data)
dates
```

```python
 df = pd.DataFrame({'user_message': messages, 'message_date': dates})
```

```python
 # Replace \u202f with normal space and convert the date
df['message_date'] = df['message_date'].str.replace('\u202f', ' ')
df['message_date'] = pd.to_datetime(df['message_date'], format='%m/%d/%y, %I:%M %p - ')

# Rename the column
df.rename(columns={'message_date': 'date'}, inplace=True)
```

```python
users = []
messages = []

for message in df['user_message']:
    entry = re.split(r'([\w\W]+?):\s', message)
    if entry[1:]:  # user name exists
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])
```

```python
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
```

```python
 df['only_date'] = df['date'].dt.date
df['year'] = df['date'].dt.year
```
